```python
import pandas as pd
import re
```

```python
class NewEmployee():
  def __init__(self, dataframe):
    self.dataframe = dataframe

  def add_details(self):
    print("Enter Employee ID")
    while True:
      try:
        employee_id = int(input())
        if employee_id in self.dataframe["Employee ID"].values:
          print("Employee ID already exists")
          continue
      except ValueError:
        print("Invalid Input, ID must be integer")
        continue
      else:
        break

    print("Enter Employee Name")
    while True:
      employee_name = str(input())
      pattern = "^[a-zA-Z ]+$"
      if (re.search(pattern, employee_name)):
        break
      else:
        print("Invalid Input, Name must be only letters")
        continue

    # If adding occupation, can match occupation from a set

    new_row = [employee_id, employee_name]
    self.dataframe.loc[len(self.dataframe.index)] = new_row

    return self.dataframe
```

```python
class ExEmployee():
  def __init__(self, dataframe):
    self.dataframe = dataframe

  def remove_details(self):
    print("Enter Employee ID")

    a = int(input())
    if a not in self.dataframe["Employee ID"].values:
      print("Employee ID does not exist")
      return self.dataframe
    else:
      self.dataframe.drop(self.dataframe.index[self.dataframe["Employee ID"] == a], inplace = True)

    return self.dataframe
```

```python
class EmployeeSearch():
  def __init__(self, dataframe):
    self.dataframe = dataframe

  def search_details(self):
    print("Enter Employee ID or Name")
    b = input()

    if b in self.dataframe["Employee Name"].values:
      return self.dataframe[self.dataframe["Employee Name"] == b]

    try:
      y = int(b)
      if int(b) in self.dataframe["Employee ID"].values:
        return self.dataframe[self.dataframe["Employee ID"] == int(b)]
    except ValueError:
      print("Employee does not exist")
      return self.dataframe
    else:
      print("Employee does not exist")
      return self.dataframe
```

```python
class BaseFunction():
  def __init__(self, dataframe):
    self.dataframe = dataframe

  def run_base(self):
    while True:
      print("\nWhat would you like to do?")
      functions = {
          1 : "Add New Employee",
          2 : "Delete Employee",
          3 : "Search Employee",
          4 : "Update Employee",
          5 : "Exit"
          }
      print(functions)
      try:
        f = int(input())
      except ValueError:
        print("Invalid Input")
        continue

      if f == 1:
        print(functions[f], "\n")
        employee = NewEmployee(self.dataframe)
        print("\n", employee.add_details())
        continue
      elif f == 2:
        print(functions[f], "\n")
        employee = ExEmployee(self.dataframe)
        print("\n", employee.remove_details())
        continue
      elif f == 3:
        print(functions[f], "\n")
        employee = EmployeeSearch(self.dataframe)
        print("\n", employee.search_details())
        continue
      elif f == 4:
        print(functions[f], "\n")
        employee = EmployeeUpdater(self.dataframe)
        print("\n", employee.update_details())
        continue
      elif f == 5:
        break
      else:
        print("Invalid Option\n")
        continue
```

```python
class System():
  def __init__(self, dataframe):
    self.dataframe = dataframe

  def run_system(self):
    base = BaseFunction(self.dataframe)
    base.run_base()
```
